```python
import pandas as pd
dates = pd.read_csv(
    "/home/bjorge_local/project_src/spermull-bonn/data/spermullterminebonn2025.csv",
    index_col=0)
geocodes = pd.read_csv(
    "/home/bjorge_local/project_src/spermull-bonn/data/geocoded_addr.csv",
    index_col=0)
```

```python
# inner join based on postal code and street.
# This does not care about the fact that some house numbers are 
# split up in different sections (not yet, see below)

merged_df = pd.merge(geocodes, dates, on=['PLZ1', 'STRASSE1'])
merged_df.head()
```

```python
def condition(row):
    """Filter out entries whose house number does not fall within their respective range"""
    if row['HNR'] % 2 == 0:  # n is even
        return row['HNR_GE_AB'] <= row['HNR'] <= row['HNR_GE_BIS']
    else:  # n is odd
        return row['HNR_UG_AB'] <= row['HNR'] <= row['HNR_UG_BIS']

filtered_df = merged_df[merged_df.apply(condition, axis=1)]

filtered_df.set_index(["TERMIN001", "TERMIN002", "TERMIN003"]).sort_values(by=["TERMIN001", "TERMIN002", "TERMIN003"])
```

```python
# Melt the DataFrame to have a single date column
melted_df = filtered_df.melt(
    id_vars=['LAT', 'LON'], 
    value_vars=['TERMIN001', 'TERMIN002', 'TERMIN003'],
    var_name="termin_nr",
    value_name='date')

# Set MultiIndex for quick lookup
melted_df.set_index(['date', melted_df.index], inplace=True)

from isf_pandas_msgpack import to_msgpack, read_msgpack
to_msgpack(
    "/home/bjorge_local/project_src/spermull-bonn/data/geo_dates.msg", 
    melted_df)
```
